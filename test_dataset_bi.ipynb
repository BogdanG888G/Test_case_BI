# Создание модели прогноза для каждого мессяца на 2025 год

```python
dict_to_replace = {'Январь 2025 год': pd.to_datetime('2025-01'),
                   'Весь 2024 год': pd.to_datetime('2024-06'),
                   'Декабрь 2024 год': pd.to_datetime('2024-12'),
                   'Ноябрь 2024 год': pd.to_datetime('2024-11')}
over_all = over_all.replace(dict_to_replace)
display(over_all)
```

## Первая ML-модель (учитывая весь датасет)

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest
from sklearn.metrics import mean_squared_error, r2_score


over_all = pd.read_csv("Общий датасет.csv")

features = over_all.drop(columns=['Sales'])
label_encoders = {}


for col in features.columns:
    if features[col].dtype == 'object':  
        le = LabelEncoder()
        features[col] = le.fit_transform(features[col].astype(str))
        label_encoders[col] = le  

target = over_all['Sales']


best_features = SelectKBest(k=11)
best_features.fit(features, target)
selected_features = best_features.get_feature_names_out()
features = features[selected_features]


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42, shuffle=True)


model = RandomForestRegressor(n_estimators=300, criterion='squared_error', random_state=42)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


print(f'Метрика R²: {r2_score(y_test, y_pred):.4f}')
print(f'Метрика MSE: {mean_squared_error(y_test, y_pred):.2f}')

# --- Прогноз на 2025 год (по дням) ---

future_dates = pd.date_range(start='2025-02-01', end='2025-12-31', freq='D')  
future_df = pd.DataFrame({'Дата': future_dates})  

# Добавляем временные признаки
future_df['День месяца'] = future_df['Дата'].dt.day
future_df['Месяц'] = future_df['Дата'].dt.month
future_df['День недели'] = future_df['Дата'].dt.weekday  

# Добавляем название месяца в формате "Февраль 2025"
month_names = {
    1: "Январь", 2: "Февраль", 3: "Март", 4: "Апрель", 5: "Май", 6: "Июнь",
    7: "Июль", 8: "Август", 9: "Сентябрь", 10: "Октябрь", 11: "Ноябрь", 12: "Декабрь"
}
future_df['Месяц (текст)'] = future_df['Месяц'].map(month_names) + " 2025"

# Средние значения для остальных признаков
feature_means = features.mean()

# Добавляем вариации в данные
for col in features.columns:
    if col not in ['День месяца', 'Месяц', 'День недели']:
        trend = np.linspace(-0.1, 0.1, len(future_df))  
        noise = np.random.normal(0, 0.05, len(future_df))  
        future_df[col] = feature_means[col] * (1 + trend + noise)


for col in label_encoders:
    if col in future_df.columns:
        unique_values = label_encoders[col].classes_
        future_df[col] = np.random.choice(unique_values, len(future_df))
        future_df[col] = label_encoders[col].transform(future_df[col])


missing_cols = [col for col in selected_features if col not in future_df.columns]
for col in missing_cols:
    future_df[col] = features[col].mean()

future_df = future_df[selected_features]


future_df['Predicted Sales'] = model.predict(future_df[selected_features])


future_df['Дата'] = future_dates
future_df['Месяц (текст)'] = future_df['Дата'].dt.strftime('%B %Y')  # Например, "Февраль 2025"


agg_fut = future_df.groupby('Месяц (текст)')['Predicted Sales'].sum()

display(agg_fut)

```

Эта модель считается более правильной и полной с точки зрения данных, но сами предсказания очень маленькие по отдельным дням. Вероятно,стоит еще учитывать точную дату (до дня), так как она играет немаловажную роль в образовании продаж

## Вторая ML-модель (учитывая агрегацию датасета по периодам)

Эта ML-модель была обучена на основе сгруппированного по периодам датасета. Плюс заключается в том, что у нас сразу есть готовая сумма продаж за периоды и можем предсказать для будущих периодах, но минус заключается в том, что модель опирается лишь на эти данные, а это около 4 строк... 

```python
lol = over_all.groupby('Период')['Sales'].sum().iloc[1:].reset_index()


x = lol['Период']
y = lol['Sales']


le = LabelEncoder()
x_encoded = le.fit_transform(x).reshape(-1, 1)


model = RandomForestRegressor(n_estimators=400, random_state=42)
model.fit(x_encoded, y)


y_pred = model.predict(x_encoded)


r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print(f'R² модели: {r2:.4f}')
print(f'MSE модели: {mse:.2f}')


future_months = ['Февраль 2025', 'Март 2025', 'Апрель 2025', 'Май 2025', 'Июнь 2025', 
                 'Июль 2025', 'Август 2025', 'Сентябрь 2025', 'Октябрь 2025', 'Ноябрь 2025', 'Декабрь 2025']


future_months_encoded = LabelEncoder().fit_transform(future_months).reshape(-1, 1)


predictions = model.predict(future_months_encoded)

# Добавляем случайный разброс (5% от значения)
noise = np.random.normal(0, 0.05, len(predictions))  
predictions = predictions * (1 + noise)


done_df = pd.DataFrame({'Месяц': future_months, 'Продажи': predictions})
add_df = pd.DataFrame({'Месяц': ['Весь 2024 год', 'Декабрь 2024 год', 'Ноябрь 2024 год', 'Январь 2025 год'],
                       'Продажи': [301189, 58211, 45012, 85445]})

done_df = pd.concat([add_df, done_df], ignore_index=True)

print(done_df)
```

```python
done_preds = done_df.to_csv('Прогноз на 2025 год.csv')
```
